# Задача 2

```python
# 1
x = np.random.uniform(-3, 2, 10)
print("Выборка:")
print(x, "\n")


def makeConfidenceInterval(x):    
    # 2
    mean = np.mean(x)
    std = np.std(x)
    print("mean={}, Корень из дисперсии данных std={}".format(mean, std))

    # 3
    sq = len(x) ** 0.5 # Корень из длины выборки
    confidence_interval = [mean - 3 * std / sq, mean + 3 * std / sq]
    print("\nДоверительный интервал для среднего по правилу 3 сигма:")
    print(confidence_interval)

    # 4
    print("Попадает ли теор. среднее в интервал? {}".format(confidence_interval[0] <= -0.5 <= confidence_interval[1]))
    
makeConfidenceInterval(x)
```

Теоретическое среднее для равномерного распределения на отрезке $[-3; 2]$ равно $Ex = \frac{a + b}{2} = -0.5$.

Дисперсия равна $Dx = \frac{(b-a)^2}{12} = \frac{25}{12}$ 

$\sqrt{Dx} = 1.4433$

Теоретическое среднее попало в построенный интервал.


```python
def generate(eps, size=10):
    # Строим загрязненную выборку
    x1 = np.random.uniform(-3, 2, size)
    x2 = 5 * np.random.standard_cauchy(size)
    x3 = np.random.binomial(1, eps, size)
    x = x2 * x3 + x1 * (1 - x3)
    return (x, x3)
```

```python
def dirtyDistribution(eps=0.1, size=10):
    #5
    x, mask = generate(eps, size)

    #6
    print("{}Eps={}{}".format('-' * 20, eps, '-' * 20))
    print(x)
    print("{} - единицы соответсвуют значениям из распределения Коши\n".format(mask))
    makeConfidenceInterval(x)
    print("\n")
```

```python
#7
for i in range(5):
    dirtyDistribution(0.1, 10)
```

Среднее всегда попадает в интервал, построенный по правилу 3-х сигм. Интервал значительно расширяется, когда в выборке присутствует явный выброс, поэтому в этих случаях, теор. среднее для равномерного распределения все равно попадает в интервал. При небольшом зашумлении эмпирическое среднее не сильно уклоняется от теор. среднего.

```python
# Пункт 8
epsilons = np.linspace(0.1, 0.6, 6)
for eps in epsilons:
    dirtyDistribution(eps)
```

```python
def getIntervals(dist, n, **args):
    intervals = []
    for i in range(1, n):
        intervals.append(dist.ppf(float(i) / n, **args))
    return intervals


def getChiStatValue(X, intervals):
    n = len(intervals) + 1
    freq = [0 for i in range(n)]
    for x in X:
        if (x < intervals[0]):
            freq[0] += 1
        else:
            if (x >= intervals[-1]):
                freq[-1] += 1
            else:
                for i in range(1, n - 1):
                    if (intervals[i - 1] <= x < intervals[i]):
                        freq[i] += 1
    
#     print(freq)
    p = 1.0 / n
    total = len(X)
    
    chival = 0.0
    for i in range(n):
        chival += (freq[i] - total * p) ** 2 / (total * p)
    return chival

def chiSquareCrit(x, intervals, statSign):
    chival = getChiStatValue(x, intervals)
    pval = chi2.cdf(chival, len(intervals))
    print(pval)
    return pval <= statSign
```

Гипотеза $H_0$: наблюдения имеют хи-квадрат распределение с $4$ степенями свободы

```python
#1
x = np.random.chisquare(4, 12)
chiints = getIntervals(chi2, 4, df=4)
#2 
if chiSquareCrit(x, chiints, 0.95):
    print("Принимаем гипотезу H0")
else:
    print("Отвергаем гипотезу H0")
```

Гипотеза $H_1$: наблюдения имеют распределение Рэлея с параметром $a = 3$ 

Гипотеза $H_2$: наблюдения имеют показательное распределение с параметром $\lambda = 0.25$

```python
#3
# генерируем равноверояные интервалы для распределения Рэлея и экспоненциального распределения
rayleighInts = getIntervals(rayleigh, 4, scale = 3)
expInts = getIntervals(expon, 4, scale=1.0/0.25)

if chiSquareCrit(x, rayleighInts, 0.9):
    print("Принимаем гипотезу H1")
else:
    print("Отвергаем гипотезу H1")
    
if chiSquareCrit(x, expInts, 0.9):
    print("Принимаем гипотезу H2")
else:
    print("Отвергаем гипотезу H2")
```

```python
#4
baseN = 2
i = 0
flag = True

# Последовательно удваивая добива
while (flag):
    size = baseN * 2 ** i
    print("-----------i={}----Кол-во наблюдений={}---------".format(i, size))
    x = np.random.chisquare(4, size)
    print("Принимаем H0? {}".format(chiSquareCrit(x, chiints, 0.95)))
    i += 1
    
    f1 = chiSquareCrit(x, rayleighInts, 0.9)
    print("Принимаем H1? {}".format(f1))
    f2 = chiSquareCrit(x, expInts, 0.9)
    print("Принимаем H2? {}".format(f2))
    #flag = f1 or f2
    flag = i < 12
    
print(i, baseN * 2 ** i)
```

Видим, что критерий хи-квадрат набирает достаточную статистическую силу при размере выборки в 100 элементов (при разбиении на 4 равновероятных интервала тестируемых распределений). Таким образом, для успешного применения данного критерия у нас должно быть много данных, иначе некоторые распределения будут неотличимы друг от друга.
